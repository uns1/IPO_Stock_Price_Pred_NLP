```python
def clean_tweet(t):

    t = t.replace('RT ','')
    t = t.replace('$','')
    t = t.replace('#','')
    
    for word in t:
        if word.startswith(("@")) == True:
            t = t.replace(word,'')
            
    t = re.sub(r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))''','',t)

    t = t.replace(',','')
    t = t.replace('.','')
    t = t.replace(';','')
    t = t.replace("'",'')
    t = t.replace('"','')
    t = t.replace('...','')
    
    return(t.lower())
```
